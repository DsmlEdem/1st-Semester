```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n=44 # μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
p = 0
REG = nx.watts_strogatz_graph(n,d, p)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n = 44 
M = 150
# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M,seed=599)
# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n = 44
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n,R)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, "pos") # get position of nodes

# Plot RGG
plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    node_size=80,
    cmap=plt.cm.Reds_r,
)
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n = 44
d = 4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n,d)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n = 44
d = 4
p = 0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n,d,p)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
from statistics import variance 

def cc(G):    
    # Β1.1
    clustering_coeff = nx.clustering(G)
    _ = plt.hist(clustering_coeff.values(),edgecolor='black', bins='auto')
    plt.title("Histogram of clustering coefficient")
    plt.show()
    print("Clustering coefficient of every node", clustering_coeff)
    # Β1.2
    avg_clustering_coeff = nx.average_clustering(G)
    print("Average clustering coefficient is", avg_clustering_coeff)
    var_clustering_coeff = variance(clustering_coeff.values())
    print("Variance of clustering coefficient is", var_clustering_coeff)
    # Β1.3
    min_clustering_coeff = min(clustering_coeff.values())
    min_keys = [k for k, v in clustering_coeff.items() if v == min_clustering_coeff]
    max_clustering_coeff = max(clustering_coeff.values())
    max_keys = [k for k, v in clustering_coeff.items() if v == max_clustering_coeff]
    print("Clustering coefficient value is",min_clustering_coeff,"and nodes with minimum clustering coefficient are", min_keys)
    print("Clustering coefficient value is",max_clustering_coeff,"and nodes with maximum clustering coefficient are", max_keys)
```

```python
print("Regular Network:")
cc(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
cc(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
cc(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
cc(SF)
print("-----------------------------------------------------------------")
print("Small World:")
cc(SW)
print("-----------------------------------------------------------------")
```

**REG:** Ο συντελεστής ομαδοποίησης είναι ίδιος για όλους τους κόμβους κάτι που είναι εύκολα προλέψιμο λόγο της ομοιομορφίας του συγκεκριμένου δικτύου. Συνεπώς ο μέσος όρος είναι ο ίδιος ο συντελεστής ομαδοποίησης και η διακύμναση είναι 0 καθώς δεν υπαρχει κόμβος που να μην ακολουθεί το μοτίβο που σχηματίζεται.

**RGER:** Ο μέσος συντελεστής ομαδοποίησης είναι ο μικρότερός που παρατηρήθηκε καθώς το δίκτυο παράγεται τυχαία. Παρόλο που ο αριθμός των συνδέσεων είναι αρκετά μεγάλος δεν είναι αρκετό ώστε να παραχθούν πολλές κλίκες σε αυτό τον RGER γράφο. Αυτό αντικατροπτίζεται και από την μέση τιμή. Το εύρος κυμμαίνεται από 0 (όπου αρκετοί κόμβοι βρίσκονται κοντά σε αυτή την τιμή συντελεστή) μέχρι 0.38 (μόνο ένας κομβος έχει αυτή την τιμή) που απέχει από την μέση τιμή. 

**RGG:** Σε αντίθεση με τον προηγούμενο, αυτός ο γράφος έχει υψηλό μέσο συντελεστή ομαδοποίησης και μικρή διακύμανση, που σημαίνει ότι γενικά παράγονται αρκετές κλίκες καθώς τα σημεία συνδέονται με βάση την γεωμετρική τους απόσταση. Αν το R ήταν μικρότερο, τότε θα είχαμε και μικρότερο συντελεστή ομαδοποίησης. Το εύρος είναι από 0.5 εως 1 και οι τιμές συσσωρεύονται γύρω απο την μέση τιμή με μια φαίνομενικά πτωτική τάση. 

**SF:** Περιμένουμε πως ο μέσος συντελεστής ομαδοποίησης ενός SF δικτύου θα είναι μεγαλύτερος από ότι σε ένα RG δίκτυο, κάτι που ισχύει στην συγκεκριμένη περίπτωση. Ο συντελεστής κυμμαίνεται από 0 ως 0.5 ενώ η μέση τιμή βρίσκεται περίπου στην μέση του διαστήματος αυτού.

**SW:** Αυτό το δίκτυο έχει επίσης σχετικά μεγαλύτερο μέσο συντελεστή ομαδοποίησης κάτι που ήταν αναμενόμενο αφού σχηματίζονται αρκετά τρίγωνα σε αυτή την τοπολογία. Το εύρος ξεκινάει από 0 και αρκετοί κόμβοι έχουν συντελεστή κάτω από 0.4 αλλά υπάρχουν και σποραδικά μερικοί κόμβοι που αποκλίνουν απο αυτή την τάση, με αποτέλεσμα η μέγιστη τιμή συντελεστή συσχέτισης να είναι το 1.


### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
from statistics import mean 

def compute_path_lengths(G):
    # B2.1
    all_shotest_paths = dict(nx.all_pairs_shortest_path_length(G))
    shortest_path_list = []
    # Κρατάω μόνο τις μη μηδενικές τιμές
    [shortest_path_list.append(v2) for k,v in all_shotest_paths.items() for v2 in v.values() if v2!=0]    
    avg_shortest_path = mean(shortest_path_list)
    print("Average shortest path is", avg_shortest_path)
    var_shortest_path = variance(shortest_path_list)
    print("Variance of shortest path is", var_shortest_path)
    # B2.2
    diam = nx.diameter(G)
    print("Diameter of the graph is", diam)
    rad = nx.radius(G)
    print("Radius of the graph is", rad)    
```

```python
print("Regular Network:")
compute_path_lengths(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
compute_path_lengths(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
compute_path_lengths(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
compute_path_lengths(SF)
print("-----------------------------------------------------------------")
print("Small World:")
compute_path_lengths(SW)
print("-----------------------------------------------------------------")
```

**REG:** Το δίκτυο με τις μεγαλύτερες τιμές. Οι τιμές της διαμέτρου και της ακτίνα του είναι ίσες, κάτι που φαίνεται λογικό για ένα REG δίκτυο. Επιπλέον, η διακύμανση είναι αρκετά υψηλή αφού το δίκτυο δημιουργήθηκε με βαθμό κόμβων 4 και η απόσταση μεταξύ κόμβων είναι μεγάλη. Αυτά τα αποτελέσματα συνάδουν με το γεγονός πως κάθε κόμβος έχει τον ίδιο βαθμό (degree) και ότι η τοπολογία του είναι ομοιόμορφη για κάθε κόμβο. 

**RGER:** Έχει τις δεύτερες μικρότερες τιμές στο μέσο μήκος μονοπατιού και την διακύμανση. Το σχήμα του επιτρέπει τον σχηματισμό σχετικά μέτριου μήκους μονοπατιών. Η διάμετρος και η ακτίνα του έχουν περίπου το ίδιο μήκος.

**RGG:** Συνήθως έχουν μεγάλο μήκος μέσου μονοπατιού αλλά αυτό δεν ισχύει σ' αυτή την περίπτωση συγκριτικά με τα υπόλοιπα γραφήματα. Το συγκεκριμένο γράφημα έχει σχετικά μεγάλη διακύμανση, ενώ η διάμετρος είναι διπλάσια από την ακτίνα του. Αυτό το γράφημα ειναι χωρικό (spacial) οπότε οπτικά φανερώνονται κάποια από τα χαρακτηριστικά του.  

**SF:** Το μικρότερο μέσο μήκος μονοπατιού κάτι που εξηγείται και από την θεωρία του μαθήματος. Πολύ λίγοι κόμβοι στα SF δίκτυα έχουν πολύ μεγάλο βαθμό, ενω ο κανόνας είναι το ακριβώς αντίθετο. Επίσης έχει την μικρότερη διάμετρο και ακτίνα. 

**SW:** Θα περιμέναμε κι αυτό το γράφημα να έχει μικρό μέσο μήκος μονοπατιού αλλά αυτό δεν ισχύει επίσης συγκριτικά με τα υπόλοιπα μέσα μήκη των συγκεκριμένων στιγμοτύπων των δικτύων. Η διάμετρος είναι πάλι μεγαλύτερη από την ακτίνα ενώ συγκριτικά με τα υπόλοιπα έχει μεγάλες τιμές. 

Η διάμετρος δηλώνει το μέγιστο από τα μέγιστα ελάχιστα μήκη μονοπατιών (εκκεντρότητα) ενώ η ακτίνα το ελάχιστο από τις εκκεντρότητες. Μ' αυτό τον τρόπο φράσουμε την εκκεντρότητα και υπάρχει ένα εύρος που παρατηρείται με αυτό τον τρόπο.


#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    all_shotest_paths = dict(nx.all_pairs_shortest_path_length(G))
    diam = nx.diameter(G)
    node_and_connected_nodes_dict = {}
    number_of_connected_nodes = []
    # Κρατάω τους κόμβους με ελάχιστο μήκος μονοπατιού ίσο με την διάμετρο
    for k, v in all_shotest_paths.items():
      connected_nodes = []
      for k2,v2 in v.items():
        if v2 == diam:
          connected_nodes.append(k2)
      node_and_connected_nodes_dict[k] = connected_nodes
      number_of_connected_nodes.append(len(connected_nodes))
    # Βρίσκω το μέγιστο αριθμό συνεδεμένων κόμβων και αναζητώ όλους κόμβους που έχουν αυτό τον αριθμό συνεδεμένων κόμβων
    max_node_num = max(number_of_connected_nodes)
    max_node = []
    node_and_connected_nodes_dict_copy = node_and_connected_nodes_dict.copy()
    [node_and_connected_nodes_dict_copy.pop(k) for k, v in node_and_connected_nodes_dict.items() if len(v)!=max_node_num]
    print("List of nodes that are connected with the most nodes having shortest path length equal to the diameter of the graph ")
    for k,v in node_and_connected_nodes_dict_copy.items():
      print("The node is",k,"connected with nodes",v)
```

```python
print("Regular Network:")
find_nodes(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
find_nodes(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
find_nodes(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
find_nodes(SF)
print("-----------------------------------------------------------------")
print("Small World:")
find_nodes(SW)
print("-----------------------------------------------------------------")
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):  
    ecc = nx.eccentricity(G)
    diam = nx.diameter(G)
    periphery = []  
    [periphery.append(k) for k,v in ecc.items() if v==diam]
    print("The periphery of the graph is",periphery)
```

```python
print("Regular Network:")
find_periphery(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
find_periphery(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
find_periphery(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
find_periphery(SF)
print("-----------------------------------------------------------------")
print("Small World:")
find_periphery(SW)
print("-----------------------------------------------------------------")
```

```python
def find_center(G):
    ecc = nx.eccentricity(G)
    rad = nx.radius(G)
    center = []  
    [center.append(k) for k,v in ecc.items() if v==rad]
    print("The center of the graph is",center)
```

```python
print("Regular Network:")
find_center(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
find_center(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
find_center(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
find_center(SF)
print("-----------------------------------------------------------------")
print("Small World:")
find_center(SW)
print("-----------------------------------------------------------------")
```

**RGER:** Οι περισσότεροι κόμβοι του RGER ανήκουν στην διάμετρο του ενώ οι υπόλοιποι στην περίμετρο του. Φαίνεται να συνάδει με το γεγονός πως η διαφορά της διαμέσου με την ακτίνα ήταν 1 και πως το μέσο ελάχιστο μονοπάτι ήταν 2. 

**RGG:** Αυτή η τοπολογία έχει πολύ λιγότερους κόμβους σε σχέση με τις προηγούμενες στην περίμετρο και την διάμετρό.

**SF:** Επίσης διατηρείται το μικρό ποσοστό κόμβων όπως και στο RGG. Η περίμετρος του αποτελείται από δύο μόνο κόμβους ενώ το κέντρο του από έναν περισσότερoι σε σχέση με το RGG.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
import warnings
warnings.filterwarnings('ignore')

def find_katz(G):
    # Γ4.1
    a = 1/max(nx.adjacency_spectrum(G)) - 0.01
    b = 1
    # Παίρνω μονο το αληθινό μέρος
    katz_centr = [i.real for i in list(nx.katz_centrality(G,alpha=a,beta=b).values())]
    _ = plt.hist(katz_centr,edgecolor='black', bins='auto')
    plt.title("Histogram of katz centrality")
    plt.show()
    # Γ4.2
    # Υπολογίζω customly την μέση τιμη και διακύμανση λόγω ασυμβατότητας των συναρτήσεων με μιγαδικούς αριθμούς 
    katz_centr_np = np.asarray(katz_centr)
    mean_katz = np.sum(katz_centr_np)/len(katz_centr)
    variance_katz = np.sum(np.abs(katz_centr_np - mean_katz)**2)/len(katz_centr)
    print("The mean katz centrality value of the graph is ",mean_katz, "and the variance is",variance_katz)
```

```python
print("Regular Network:")
find_katz(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
find_katz(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
find_katz(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
find_katz(SF)
print("-----------------------------------------------------------------")
print("Small World:")
find_katz(SW)
print("-----------------------------------------------------------------")
```

**REG:** Από την θεωρία είναι αναμενόμενο πως η κεντρικότητα των REG δικτύων είναι σταθερή, εξού και το ιστόγραμμα αλλά και  διακύμανση που είναι σχεδόν ίση με 0. Και σε αυτή την περίπτωση έχει την μεγαλύτερη τιμή μέσης κεντρικότητας σε σχέση με τα υπόλοιπα δίκτυα.

**RGER:** Οι τιμές της κεντρικότητας katz των κόμβων φαίνεται να κατανέμονται ομοιόμορφα γύρω από την μέση τιμή με μια μικρή απόκλιση.

**RGG:** Η κατανομή των κόμβων φαίνεται παρόμοια με το RGER δίκτυο αλλά με περισσότερες αυξομειώσεις και χωρίς ιδιαίτερη απόκλιση.

**SF:** Όπως και στις περισσότερες κατανομές κεντρικότητας για τα SF, οι τιμές συγκεντρώνονται χαμηλά ενώ οι υψηλές τιμές κεντρικότητας είναι πολύ λίγες. 

**SW:** Οι τιμές συγκετρώνονται γύρω από την μέση τιμή δημιουργώντας καμπύλη. 

Η κεντρικότητα katz λαμβάνει υπόψη όλους τους δυνατούς περίπατους μεταξύ των κόμβων.

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    degr_centr = list(nx.degree_centrality(G).values())
    max_degr_centr = max(degr_centr)
    max_node_degr_centr = [i for i in range(len(degr_centr)) if degr_centr[i]==max_degr_centr]
    closs_centr = list(nx.closeness_centrality(G).values())
    max_closs_centr = max(closs_centr)
    max_node_closs_centr = [i for i in range(len(closs_centr)) if closs_centr[i]==max_closs_centr]
    betw_centr = [round(i,14) for i in list(nx.betweenness_centrality(G).values())]
    max_betw_centr = max(betw_centr)
    max_node_betw_centr = [i for i in range(len(betw_centr)) if betw_centr[i]==max_betw_centr]
    a = 1/max(nx.adjacency_spectrum(G)) - 0.01
    b = 1
    katz_centr = [i.real for i in list(nx.katz_centrality(G,alpha=a,beta=1).values())]
    max_katz_centr = max(katz_centr)
    max_node_katz_centr = [i for i in range(len(katz_centr)) if katz_centr[i]==max_katz_centr]
    print("The most central nodes of the graph:")
    print("based on degree centrality",max_node_degr_centr)
    print("based on clossness centrality",max_node_closs_centr)
    print("based on betweeness centrality",max_node_betw_centr)
    print("based on katz centrality",max_node_katz_centr)
```

```python
print("Regular Network:")
find_central(REG)
print("-----------------------------------------------------------------")
print("Random Network Erdos-Renyi:")
find_central(RGER)
print("-----------------------------------------------------------------")
print("Random Geometrical Graph:")
find_central(RGG)
print("-----------------------------------------------------------------")
print("Scale-Free:")
find_central(SF)
print("-----------------------------------------------------------------")
print("Small World:")
find_central(SW)
print("-----------------------------------------------------------------")
```

**REG:** Κάθε κόμβος θεωρείται "κεντρικός" αφού έχει την ίδια τιμή κεντρικότητας με τους υπόλοιπους. Το αποτέλεσμα είναι λογικό αφού δεν υπάρχει διαφορά στις συνδέσεις των κόμβων του δικτύου.

**RGER:** Σ' αυτό το δίκτυο έχει δημιουργηθεί ένας κόμβος που έχει τις μεγαλύτερες κεντρικότητες από όλους τους άλλους. 

**RGG:** Παρόμοια με το προηγούμενο, υπάρχει ένας κόμβος με τις μεγαλύτερες τιμές κεντρικότητας, με εξαίρεση την κεντρικότητα του βαθμού του κόμβου όπου υπάρχει ένας ακόμα κόμβος με ίδια τιμή.

**SF:** Υπάρχουν δύο κόμβοι που έχουν μονοπωλήσει σε διάφορους συνδιασμούς την "κεντρικότητα". 

**SW:** Στην τελευταία περίπτωση εμφανίζονται μια ποικιλία "κεντρικών" κόμβων που σε καμία περίπτωση δεν συγκρίνονται με το REG αλλά είναι αρκετοί σε σχέση με τις υπόλοιπες τοπολογίες. Οι περισσότεροι από αυτούς εμφανίζονται ως οι κεντρικότεροι σε δύο από τις τέσσερεις μετρικές κεντρικότητας.

```python
print("Connectivity SW") 
d = [i for i in range(2,11,2)]
n = [i for i in range(100,196,20)]
p = [i*0.1 for i in range(1,8,2)]
random.seed(seedno)
SF_list = []
# Για κάθε τιμή του d και του p αποθηκεύω διαφορετικά στατιστικά
num_connected = np.zeros((len(d),len(p))).tolist()
num_tot = np.zeros((len(d),len(p))).tolist()
# Δημιουργία γράφων με διάφορες παραμέτρους
for n_i in n:
    for i in range(len(d)):
      for j in range(len(p)):
        new_gr = nx.watts_strogatz_graph(n_i,d[i],p[j])
        if nx.is_connected(new_gr):
          num_connected[i][j] +=1
        SF_list.append(new_gr)
        num_tot[i][j]+=1
precentage1 = np.sum(np.asarray(num_connected),axis = 0) /np.sum(np.asarray(num_tot),axis = 0) 
precentage2 = np.sum(np.asarray(num_connected),axis = 1) /np.sum(np.asarray(num_tot),axis = 1)
plt.bar([str(round(i,1)) for i in p],precentage1)
plt.title("Bar plot of connectivity based on p value")
plt.show()
plt.bar(d,precentage2)
plt.title("Bar plot of connectivity based on d value")
plt.show()
```
