```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n = 48 # μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d = 4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n, d, 0)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n = 48
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n = 48
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
def draw_rgg(g):
    pos = nx.get_node_attributes(g, "pos")

    dmin = 1
    ncenter = 0
    for n in pos:
        x, y = pos[n]
        d = (x - 0.5)**2 + (y - 0.5)**2
        if d < dmin:
            ncenter = n
            dmin = d

    p = dict(nx.single_source_shortest_path_length(g, ncenter))

    plt.figure(figsize=(8,8))
    nx.draw_networkx_edges(g, pos, alpha=0.4)
    nx.draw_networkx_nodes(g, pos, nodelist=list(p.keys()), node_size=80, node_color=list(p.values()), cmap=plt.cm.Reds_r)
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.axis("off")
    plt.show()

draw_rgg(RGG)
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n = 48
d = 4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n, d, seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)

print(nx.periphery(SF))
print(nx.center(SF))
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n = 48
d = 4
p = 0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n, d, p)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)

# print(nx.info(SW))
# print(nx.clustering(SW))
# print(type(nx.clustering(SW)))
```

#### Β1.1 Υπολογισμός και κατανομή clustering coefficients.

```python
graphs = {"REG": REG, "RGER": RGER, "RGG": RGG, "SF": SF, "SW": SW}
ccs = {gname: nx.clustering(g) for gname, g in graphs.items()}

fig = plt.figure()
for i, (gname, cc) in enumerate(ccs.items()):
    ax = fig.add_subplot(2, 3, i+1)
    ax.hist(cc.values())
    ax.set_title(gname)
    ax.set_xlabel("Clustering Coefficient")
    ax.set_ylabel("Number of nodes")
    ax.set_xlim(-0.05, 1.05)
fig.tight_layout()
```

#### Β1.2 Υπολογισμός mean και variance των clustering coefficients.

```python
topologies = {"REG": "πεπερασμένο πλέγμα",
              "RGER": "τυχαίος γράφος Erdos-Renyi",
              "RGG": "τυχαίος γεωμετρικός γράφος",
              "SF": "scale-free δίκτυο",
              "SW": "small-world δίκτυο"
             }

for (gname, cc) in ccs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    print(f"\tΜέσος συντελεστής ομαδοποίησης: {np.mean(list(cc.values()))}")
    print(f"\tΔιασπορά συντελεστή ομαδοποίησης: {np.var(list(cc.values()))}")
```

#### Β.1.3 Εύρεση κόμβων με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης.

```python
for (gname, cc) in ccs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    
    node_min = np.argmin(list(cc.values()))
    min_cc = cc[node_min]
    node_max = np.argmax(list(cc.values()))
    max_cc = cc[node_max]
    
    print(f"\tΚόμβος με ελάχιστο συντελεστή ομαδοποίησης: {node_min}, με συντελεστή: {min_cc}")
    print(f"\tΚόμβος με μέγιστο συντελεστή ομαδοποίησης: {node_max}, με συντελεστή: {max_cc}")
```

#### Σχολιασμός

Εν πολλοίς, τα αποτελέσματα είναι τα αναμενόμενα. Ας τα δούμε, όμως, ανά τοπολογία:

##### Πεπερασμένο πλέγμα

Η πιο απλή περίπτωση. Όπως ακριβώς θα περίμενε κανείς, ο συντελεστής ομαδοποίησης είναι ακριβώς 0.5 για όλους τους κόμβους, αφού για $d=4$ ο καθένας συνδέεται με 4 γείτονές του, μεταξύ των οποίων υπάρχουν ακριβώς 3 από τις 6 συνδέσεις (αριστερός γείτονας με δεξιό γείτονα, αριστερός με αριστερότερο, δεξιός με δεξιότερο). Κατά συνέπεια, αφού ο συντελεστής ομαδοποίησης, έχει μέση, ελάχιστη και μέγιστη τιμή αυτήν, και διασπορά 0.

##### Τυχαίος γράφος Erdos-Renyi

Βλέπουμε μικρούς συντελεστές ομαδοποίησης. Συγκεκριμένα, δεν ξεπερνούν το $1/3$, έχουν μέσο περίπου $12.3\%$ και σχετικά μικρή διασπορά.

##### Τυχαίος Γεωμετρικός Γράφος

Εδώ βλέπουμε μεγάλες τιμές συντελεστών ομαδοποίησης, που είναι όλες πάνω από το 0.5, με μέσο το 0.73 περίπου, και σχετικά ομοιόμορφα κατανεμημένες.

Δεδομένου ότι τα σημεία που παράγουμε είναι εντός του τετραγώνου $[0,1] \times [0,1]$ και η ακτίνα που έχουμε θέσει είναι 0.35, καταλαβαίνουμε ότι είναι μία αρκετά μεγάλη τιμή, εξ ου και το ότι για κάθε κόμβο πολλοί γείτονές του συνδέονται μεταξύ τους.

##### Scale-free

Τιμές συντελεστών συγκεντρωμένες κυρίως κάτω από το 0.5, με σχετικά λιγότερους κόμβους να έχουν τις μεγαλύτερες τιμές, λίγο παρόμοια με την κατανομή των βαθμών των γραφημάτων αυτών δηλαδή.

##### Small-world

Κοντά στο scale-free όσον αφορά τα βαθμωτά μέτρα, δηλαδή μέση τιμή, διασπορά, μέγιστη και ελάχιστη. Και είναι λογικό, βέβαια, αφού το scale-free είναι υποπερίπτωση των small-world δικτύων. Από εκεί και πέρα, το small-world φαίνεται να έχει λίγο μεγαλύτερους συντελεστές ομαδοποίησης, και αρκετά λιγότερο συγκεντρωμένες τιμές. Που είναι λογικό, αφού τα απλά small-world δίκτυα είναι κάπως πιο decentralized, ενώ τα scale-free έχουν λίγους κεντρικούς κόμβους και παρουσιάζουν φαινόμενα coalitions κλπ.

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    lengths = nx.all_pairs_shortest_path_length(G)
    ret = []
    for u, ss_lengths in lengths:
        ss_lengths_to_others = {k:v for k, v in ss_lengths.items() if k != u}
        ret.extend(ss_lengths_to_others.values())
    return np.array(ret)

for (gname, graph) in graphs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    all_path_lengths = compute_path_lengths(graph)
    print(f"\tΤο μέσο μήκος μονοπατιού είναι {np.mean(all_path_lengths)}.")
    print(f"\tΗ διασπορά του είναι {np.var(all_path_lengths)}.")
    print(f"\tΗ διάμετρος του γραφήματος ισούται με {nx.diameter(graph)}.")
    print(f"\tΗ ακτίνα του γραφήματος ισούται με {nx.radius(graph)}.")
```

Από την άλλη, όμως, ας μην ξεχνάμε ότι εδώ η τιμή της ακτίνας είναι αρκετά μεγάλη, με αποτέλεσμα να έχουμε γενικά κάπως μεγάλη συνδεσιμότητα στον γράφο. Σκεφτείτε ότι με την ακτίνα στο 0.35, ένας και μόνο κόμβος ο οποίος να βρίσκεται ακόμα και τέρμα γωνία θα του αντιστοιχεί εμβαδό σύνδεσης $\frac{1}{4} \pi 0.35^2 \approx 0.096$, σχεδόν 1/10, δηλαδή, του συνολικού τετραγώνου!

###### Πλέγμα

Ειδικά για το πλέγμα, μπορούμε να επαληθεύσουμε λίγο και θεωρητικά τα αποτελέσματα.

Όλοι οι κόμβοι ισότιμοι, έχουν την ίδια εκκεντρότητα, η οποία μπορεί κανείς να δει εύκολα ότι είναι το μήκος μέχρι μονοπατιού που απαιτείται για να φτάσουν στον "απέναντι" κόμβο του κύκλου. Οπότε, 24 κόμβοι μακριά, με βήμα το πολύ 2 που μπορώ να κάνω κάθε φορά, στον παραδιπλανό μου, όντως η εκκεντρότητα όλων (άρα, και η ακτίνα και η διάμετρος) πρέπει να είναι 12.

Όσον αφορά την κατανομή των (ελαχίστων) μονοπατιών, είναι και αυτή ίδια για κάθε κόμβο ξεχωριστά, άρα και για όλους μαζί, και, αν το σκεφτούμε, έχει μέση τιμή περίπου το 6 και τυπική απόκλιση γύρω στο 3, γιατί πάει ομοιόμορφα σχεδόν από 0 έως και 12.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    diameter = nx.diameter(G)
    lengths = nx.all_pairs_shortest_path_length(G)
    
    first_source, dists = next(lengths)
    goal_node = first_source
    furthest_nodes = {v for v, dist in dists.items() if dist == diameter}
    max_node_count = len(furthest_nodes)
    for u, ss_paths in lengths:
        curr_furthest_nodes = {v for v, dist in ss_paths.items() if dist == diameter}
        if len(curr_furthest_nodes) > max_node_count:
            max_node_count = len(furthest_nodes)
            goal_node = u
            furthest_nodes = curr_furthest_nodes
    
    return goal_node, furthest_nodes
```

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο. 
for gname, graph in graphs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    u, furthest = find_nodes(graph)
    print(f"\tΟ κόμβος που συνδέεται με περισσότερους τέρμα-θεού κόμβους είναι ο {u}.")
    print(f"\tΟι απομακρυσμένοι του κόμβοι είναι οι εξής: {furthest}.")
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    diameter = nx.diameter(G)
    eccentricities = nx.eccentricity(G)
    return {u for u, ecc in eccentricities.items() if ecc == diameter}

# Note: there is also, of course, nx.periphery
```

```python
def find_center(G):
    radius = nx.radius(G)
    eccentricities = nx.eccentricity(G)
    return {u for u, ecc in eccentricities.items() if ecc == radius}

# Note: there is also, of course, nx.center
```

```python
for gname, graph in graphs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    print(f"\tΠεριφέρεια: {find_periphery(graph)}.")
    print(f"\tΚέντρο: {find_center(graph)}.")
```

#### Σχολιασμός

##### Πεπερασμένο πλέγμα

Στο πλέγμα, υπάρχει απόλυτη ομοιομορφία, όλοι οι κόμβοι "βλέπουν" την ίδια εικόνα. Συνεπώς, όλοι οι κόμβοι θα έχουν την ίδια εκκεντρότητα, για αυτό και είναι και στο κέντρο και στην περιφέρεια.

##### Τυχαίος γράφος Erdos-Renyi

Το γράφημα είναι τυχαίο, και γενικά τέτοια γραφήματα περιμένουμε να έχουν μία ομοιομορφία και στις αποστάσεις. Στην συγκεκριμένη περίπτωση, για του λόγου το αληθές, παρατηρούμε πρώτα από όλα ότι οι κόμβοι έχουν μοιραστεί σχεδόν ομοιόμορφα στο κέντρο και την περιφέρεια του γράφου.

Αλλά όχι μόνο αυτό! Παρατηρούμε από το προηγούμενο ερώτημα ότι η ακτίνα του συγκεκριμένου γραφήματος είναι 3 και η περιφέρεια 4. Αυτό, όμως, σημαίνει ότι κάθε κόμβος δεν μπορεί παρά να έχει εκκεντρότητα 3 ή 4, αφού είναι ακέραιος αριθμός! Συνεπώς, μπορούμε να πούμε με σιγουριά ότι το στο συγκεκριμένο γράφημα το κέντρο και η περιφέρεια αποτελούν μία διαμέριση του συνόλου των κόμβων σε δύο σύνολα.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    spectre = np.real(nx.adjacency_spectrum(G)) # eigenvalues are always real, because adjacency matrix is symmetric.
    lamda_max = max(spectre)
    katz = nx.katz_centrality(G, alpha=1/lamda_max - 0.01)
    return list(katz.values())

fig = plt.figure()
for i, (gname, graph) in enumerate(graphs.items()):
    ax = fig.add_subplot(2, 3, i + 1)
    katz = find_katz(graph)
    ax.hist(katz)
    ax.set_title(gname)
fig.tight_layout()

for gname, graph in graphs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    katz = find_katz(graph)
    print(f"\tΜέση κεντρικότητα Katz: {np.mean(katz)}.")
    print(f"\tΔιασπορά κεντρικότητας Katz: {np.var(katz)}.")
```

#### Σχολιασμός

Δεν βλέπουμε μεγάλες διαφορές σε σχέση με τα προηγούμενα αποτελέσματα.

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    degree = dict(G.degree)
    max_degree = max(degree.values())
    degree_central = {u for u, deg in degree.items() if deg == max_degree}
    
    closeness = nx.closeness_centrality(G)
    max_closeness = max(closeness.values())
    closeness_central = {u for u, cl in closeness.items() if cl == max_closeness}
    
    betweenness = nx.betweenness_centrality(G)
    max_betweenness = max(betweenness.values())
    betweenness_central = {u for u, bet in betweenness.items() if bet == max_betweenness}
    
    spectre = np.real(nx.adjacency_spectrum(G))
    lamda_max = max(spectre)
    katz = nx.katz_centrality(G, alpha=1/lamda_max - 0.01)
    max_katz = max(katz.values())
    katz_central = {u for u, kat in katz.items() if kat == max_katz}
    
    return {"degree": degree_central,
            "closeness": closeness_central,
            "betweenness": betweenness_central,
            "katz": katz_central
           }

for gname, graph in graphs.items():
    print(f"Τοπολογία: {topologies[gname]}")
    centrals = find_central(graph)
    for metric, nodes in centrals.items():
        print(f"\tΟι πιο κεντρικοί κόμβοι ως προς {metric} centrality είναι οι: \n{nodes}")
```

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;
