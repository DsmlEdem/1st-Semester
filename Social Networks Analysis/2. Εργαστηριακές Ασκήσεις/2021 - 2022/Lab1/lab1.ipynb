## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
seed = random.seed(seedno)
n=45 #μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG= nx.watts_strogatz_graph(n,d,0,seed)

nx.draw_circular(REG)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=45 
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER= nx.gnm_random_graph(n,M,seed)
nx.draw_circular(RGER)
# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=45
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG= nx.random_geometric_graph(n,R,seed=seed)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
nx.draw_circular(RGG)
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n=45 
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF= nx.barabasi_albert_graph(n,d,seed=seed)
nx.draw_circular(SF)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n=45
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW= nx.watts_strogatz_graph(n,d,p,seed=seed)
nx.draw_circular(SW)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
#Functions to plot the coefficient distribution. 
#inspired by https://stackoverflow.com/questions/64485434/how-to-plot-the-distribution-of-a-graphs-clustering-coefficient
def plot_distrib(G,cc):
    fig, ax1 = plt.subplots(ncols=1)
    ax1.hist(cc.values())
    ax1.set_xlabel('Clustering')
    ax1.set_ylabel('Frequency')
    plt.tight_layout()
    plt.show()
    
def cc(G):
    cc = nx.clustering(G)
    df = pd.DataFrame.from_dict(cc,orient='index', dtype='float', columns=['Coefficients'])
    plot_distrib(G,cc)
    avg = nx.average_clustering(G)
    print(f"Average clustering coefficient: {avg}")
    print(f"Variance of clustering coefficients: {df.var(axis=0)[0]}")
    #We get the minimum and the maximum coefficient from the graphs
    mincoefficient = df.min(axis=0)[0]
    maxcoefficient = df.max(axis=0)[0]
    #Here we extract the values we want and remove the NaN that pandas creates
    minNodes = df.where(df==mincoefficient)
    minNodes = minNodes[~minNodes['Coefficients'].isnull()]
    maxNodes = df.where(df==maxcoefficient)
    maxNodes = maxNodes[~maxNodes['Coefficients'].isnull()]
    
    print(f"nodes with minimum clustering coefficient: {minNodes.index}")
    print(f"nodes with maximum clustering coefficient: {maxNodes.index}")

    
#Regular
cc(REG)
#Random erdos renyi
cc(RGER)
#Random geometric
cc(RGG)
#Scale free
cc(SF)
#Smallworld
cc(SW)

"""
Για το regular graph βλεπουμε πως παντου ο συντελεστης ομαδοποιησης ειναι ο ιδιος διοτι ολοι οι κομβοι
δηλαδη και οι γειτονες τους ειναι συνδεδεμενοι με τον ιδιο βαθμο
Στο random geometric graph  φαινεται να υπαρχει μεγαλυτερη διασπορα στους συντελεστες.
Στο smallworld βλεπουμε πως 2 κομβοι εχουν υψηλο συντελεστη ομαδοποιησης ενω οι περισσοτεροι εχουν μικροτερο απο ~0.2-0.3, έτσι περίπου φαίνεται να λειτουργεί και το scale-free
"""
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    path_lengths = dict(nx.all_pairs_shortest_path_length(G))
    df = pd.DataFrame.from_dict(path_lengths,orient='index', dtype='float')
    mean = df.mean()[0]
    var = df.var()[0]
    diameter = nx.diameter(G)
    radius = nx.radius(G)
    print(f"mean path length: {mean}, variance: {var} \n diameter: {diameter} and radius: {radius}")
    pass

compute_path_lengths(REG)
'Βλεπουμε πως εχει μεγαλη διαμετρο και ακτινα το κανονικο δικτυο διοτι βασιζεται σε καποια τοπολογια, το ιδιο και το RGG'
compute_path_lengths(RGER)
'Δεν μας κανει εντυπωση στον τυχαιο γραφο πως εχει μεσο μηκος μονοπατιου 2, διοτι δημιουργουνται ακμες με uniform κατανομη'
compute_path_lengths(RGG)

compute_path_lengths(SF)
'Βλεπουμε πως επικρατει η φιλοσοφια του πλουσιου που γινεται πλουσιοτερος απο την μικρη ακτινα του δικτυου'
compute_path_lengths(SW)
'Επιβεβαιωνεται και απο την θεωρια οτι το μεσο μηκος μονοπατιου των smallworld ειναι περιπου 3 hops, οπως το παραδειγμα με τα τηλεφωνηματα/γραμματα'
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    path_lengths = dict(nx.all_pairs_shortest_path_length(G))
    df = pd.DataFrame.from_dict(path_lengths,orient='index', dtype='float')
    diameter = nx.diameter(G)
    path_lengths = df[df==diameter]
    node_connections = {} #dict
    for i in range(0,len(df)):
        counter = 0
        for j in range(0,len(df)):
            if path_lengths.loc[i,j]==diameter:
                counter+=1
            node_connections[i]=counter
    dt = pd.DataFrame.from_dict(node_connections,orient='index', columns=['length'])
    df = dt[dt!=0]
    df = df[~df['length'].isnull()]
    print(f"Κόμβος {df.idxmax(axis=0)[0]} Σύνολο κόμβων που συνδέεται {dt.loc[df.idxmax(axis=0)[0]][0]}")

find_nodes(REG)
find_nodes(RGER)
find_nodes(RGG)
find_nodes(SF)
find_nodes(SW)
"""Βλεπουμε πως δεν ξεφευγει πολυ ο αριθμος του κομβου που συνδεεται με τους περισσοτερους κομβους ισο με τη διαμετρο,
δηλαδη υπαρχει αντιστοιχια με το μεσο μηκος διαδρομης που ειδαμε στο προηγουμενο ερωτημα"""
```

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
#Function to plot the coefficient distribution. 
#inspired by https://stackoverflow.com/questions/64485434/how-to-plot-the-distribution-of-a-graphs-clustering-coefficient
def plot_distrib(G,dg):
    fig, ax1 = plt.subplots(ncols=1)
    ax1.hist(dg.values())
    ax1.set_xlabel('Degree centrality')
    ax1.set_ylabel('Frequency')
    plt.tight_layout()
    plt.show()

def find_degree(G):
    dg = nx.degree_centrality(G)
    plot_distrib(G,dg)
    df = pd.DataFrame.from_dict(dg,orient='index', dtype='float', columns=['Centrality'])
    mean = df.mean()[0]
    var = df.var()[0]
    print(f"mean degree centrality : {mean}, variance: {var}")
    return df

find_degree(REG)
find_degree(RGER)
find_degree(RGG)
find_degree(SF)
find_degree(SW)


'''Στον κανονικο γραφο όλοι οι κόμβοι έχουν την ίδια κεντρικότητα, στους τυχαίους γράφους ψιλοδιακρίνεται μία μέση τιμή κεντρικότητας σε όλους τους κόμβους,
Στο scale-free η πλειοψηφία των κόμβων δεν είναι καθόλου κεντρικοί, με λίγους "πλούσιους-κεντρικούς κόμβους. Στο smallworld βλέπουμε μία κατανομή παρόμοια με την κανονική ως προς την κεντρικότητα που ταυτόχρονα μοιάζει με τον κανονικό γράφο"'''
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def plot_distrib(G,ktz):
    fig, ax1 = plt.subplots(ncols=1)
    ax1.hist(ktz.values())
    ax1.set_xlabel('Katz Centrality')
    ax1.set_ylabel('Frequency')
    plt.tight_layout()
    plt.show()

def find_katz(G):
    lmax=np.real(np.max(nx.adjacency_spectrum(G)))
    alpha= (1/lmax) - 0.01
    katz = nx.katz_centrality(G,alpha=alpha)
    plot_distrib(G,katz)
    df = pd.DataFrame.from_dict(katz,orient='index', dtype='float', columns=['Centrality'])
    mean = df.mean()[0]
    var = df.var()[0]
    print(f"mean closeness centrality : {mean}, variance: {var}")
    return df

find_katz(REG)
find_katz(RGER)
find_katz(RGG)
find_katz(SF)
find_katz(SW)

#katz_centrality(G, alpha=0.1, beta=1.0, max_iter=1000, tol=1e-06, nstart=None, normalized=True, weight=None)
```
