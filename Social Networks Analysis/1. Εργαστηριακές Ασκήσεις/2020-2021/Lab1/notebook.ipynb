```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n=48 #μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG= nx.watts_strogatz_graph(n,k=d,seed=seedno,p=0)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG,with_labels = True)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=48
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER= nx.gnm_random_graph(n,M,seedno)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=48
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG=nx.random_geometric_graph(n,R,seed=seedno)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

pos = nx.get_node_attributes(RGG,"pos")

dmin = 1
ncenter = 0

# center of 2d graph is 0.5,0.5
for n in pos:
    x,y = pos[n]
    d = (x-0.5)**2 + (y-0.5)**2
    if d<dmin:
        dmin = d
        ncenter = n
        
p = dict(nx.shortest_path_length(RGG,ncenter))
list(p.values())
nx.draw_networkx_edges(RGG,pos,alpha=0.4)
nx.draw_networkx_nodes(RGG,pos,node_size=80,node_color=list(p.values()))
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n=48
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF= nx.barabasi_albert_graph(n,d,seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n=48
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW= nx.watts_strogatz_graph(n,d,p,seed=seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
def node_clustering_coefficient(G,name):
    clustering_coeff_dict = nx.clustering(G)
    plt.hist(clustering_coeff_dict.values(),label=name,alpha=0.4)
    plt.xlabel("Clustering Coefficient")
    plt.ylabel("Frequency")
    plt.legend()
    plt.plot()
    return clustering_coeff_dict

graph_dict = {REG:"Regular",RGER:"Erdos-Renyi",RGG:"Random Geometric Graph",SF:"Scale-Free",SW:"Small World"}
for graph,name in graph_dict.items():
    node_clust_dict = node_clustering_coefficient(graph,name)
    clustering_values = list(node_clust_dict.values())
    min_,max_,avg_,std_ = np.argmin(clustering_values), np.argmax(clustering_values),np.mean(clustering_values),np.std(clustering_values)
    print("For {} graph node {} has the minimum clustering coefficient of {:.3f}\n node {} has the maximum coefficient of {:.3f} and the average is {:.3f} and variance of {:.3f}\n".
          format(name,min_,clustering_values[min_],max_,clustering_values[max_],avg_,std_))
```

#### Σχολιασμός

Οπως είναι αναμενόμενο παρατηρούμε πως ο Regular γράφος μας που είναι ideally κατασκευασμένος (p=0)
παρουσιάζει το ίδιο clustering coefficient σε όλους τους κόμβους καθώς είναι όλοι συνδεδεμένοι με τον ίδιο τρόπο.

Για το Erdos-Renyi μοντέλο παρατηρούμε πολύ χαμηλό clustering coefficient μιας και ο γράφος κατασκευάζεται με τυχαίο τρόπο
επομένως ο αναμενόμενος αριθμών τριγώνων είναι χαμηλός.
Αυτό εξηγεί επίσης και την χαμηλή τιμή της διακύμανσης του εν λόγω γράφου
μιας και η τυχαιότητα τον διέπει ολόκληρο (χωρίς κάποια προτίμηση).

Παρακάτω, έχουμε το Random Geometric Graph έναν χωρικό (spatial) γράφο ο οποίος λόγο του τρόπου κατασκευής του που βασίζεται σε ακτίνες
γύρω απο τους κόμβους για την δημιουργία συζεύξεων παρουσιάζει πολύ μεγάλο clustering coefficient με μικρή διακύμανση.

Φυσικά αυτό επηρεάζεται εμφανώς απο την παράμετρο r κατά την κατασκευή του γράφου με μεγαλύτερες τιμές να αναμένεται να δίνουν μεγαλύτερη μέση τίμη και μικρότερη διακύμανση.

Παρακάτω έχουμε τα scale-free δίκτυα τα οποία έχουν περιοχές με μεγάλη πυκνότητα γύρω απο τους κόμβους hubs και μικρή πυκνότητα σε γενικές γραμμές.
Σε αυτά τα δίκτυα έχουμε πολλούς με λίγες συνδέσεις και λίγους με πάρα πολλές. Αυτό εξηγεί τόσο την μεγάλη διακύμανση του clustering coefficient καθώς και την μέση τιμή
αυτού αφού όπως φαίνεται και απο το ιστόγραμμα έχουμε κόμβους με μεγάλο cc (άρα πολλά τρίγωνα σχηματίζονται με αυτούς) και άλλους με πολύ μικρό cc μιας και δεν έχουν συνδέσεις.

Τέλος έχουμε τα small world δίκτυα στα οποία (εμπειρικά) το average pathing ~ ln(n) με το n να είναι ο αριθμός τον κόμβων. Μέσω αυτού μπορούμε να καταλάβουμε ότι υπάρχει
εύκολη συνδεσιμότητα απ\`όλους προς όλους. Αυτό ακριβώς παρατηρείται και απο το ιστόγραμμα στο οποίο βλέπουμε μια υψηλή μέση τιμή, μια σχετικά μεγάλη διακύμανση (καθώς τόσο ο αριθμός κόμβων είναι μικρός
και ίσως το p οχι βέλτιστο για την επίδειξη των small-world ιδιοτήτων) αλλά δεν βλέπουμε ακραίες τιμές προς τα επάνω που να δείχνουν ότι κάποιος κόμβος κυριαρχεί στο δίκτυο.


### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
import functools
def compute_path_lengths(G,name):
    # Create generator of all shortest paths
    shortest_path_generator = nx.all_pairs_shortest_path_length(G)
    # counts current shortest paths' distances
    sum_ = 0
    # counts number of pairs 
    pairs = 0
    #mean of shortest path
    means = []
    # idx, {dict of pair:distance}
    for _,node_pairs_dict in enumerate(shortest_path_generator):
        _,node = node_pairs_dict
        N = len(node)
        pairs += N
        distances = list(node.values())
        current_sum_of_pairs = functools.reduce(lambda a, b: a+b, distances)
        means.append(current_sum_of_pairs/N)
        sum_ += current_sum_of_pairs
    print("Mean path length for {} is {:.3f} with std {:.3f}, diameter {} and radius {}".format(name, sum_/pairs,np.std(means), nx.diameter(G), nx.radius(G)))
    
for graph,name in graph_dict.items():
    compute_path_lengths(graph,name)
```

#### Σχολιασμός

Τα *regular* δίκτυα όπως είναι αναμενόμενο έχουν μεγαλύτερο μέσο μήκος μονπατιού μιας και δεν υπάρχουν σύντομες συνδέσεις (shortcuts) μεταξύ κόμβων
που να βοηθούν στην μείωση του avg. path length.

Απο την άλλη πλευρά (στα random δίκτυα) παρατηρούμε πολύ μικρότερο μέσο μήκος μονπατιού με κυρίαρχο να είναι το scale-free.

Αυτό φυσικά δεν σημαίνει οτι εν γένει τα scale-free δίκτυα έχουν αναγκαστικά μικρότερο μέσο μήκος μονοπατιού από τα υπόλοιπα
δίκτυα καθώς εξαρτάται απο πολλούς παράγοντες όπως το μέγεθος του δικτύου ή πιο συγκεκριμένες παραμέτρους του δικτύου
όπως η πιθανότητα επανασύνδεσης στα small-world δικτυα ή το radius στο RGG ή τέλος το γ στο SF δίκτυο.


Αναφορικά με τις διαμέτρους και τις ακτίνες βλέπουμε μια παρόμοια εικόνα REG > SW > RGG > RG > SF. Αυτό συμβαίνει
καθώς εξ ορισμού η διάμετρος είναι το μέγιστο γεωδεσικό μονοπάτι του γράφου.

Είναι προφανές ότι στο REG αυτή η ποσότητα
είναι συγκριτικά μεγαλύτερη από τα υπόλοιπα δίκτυα όπως το RG που βλέπουμε οτι προσεγγίζει το βέλτιστο SF (οπού η παρουσία των κόμβων hubs βοηθάει).

Στην περίπτωση της ακτίνας βλέπουμε πως στο REG δίκτυο υπάρχει ισότητα με την διάμετρο καθώς όλες οι μέγιστες αποστάσεις μεταξύ κόμβων είναι ίσες.

Στα τυχαία δίκτυα βλέπουμε σημαντικότερες διαφορές μεταξύ της ακτίνας και της διαμέτρου με την μεγαλύτερη στο RGG καθώς εκεί όπως αναμέται
κόμβοι που είναι πιο κεντρικοί έχουν μικρότερα μέγιστα γεωδεσικά μονοπάτια.


#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G,name):
    shortest_path_generator = nx.all_pairs_shortest_path_length(G)
    max_node = -1
    max_ = -1
    diam = nx.diameter(G)
    for node,node_pairs_dict in shortest_path_generator:
        count=0
        for _,val in node_pairs_dict.items():
            if val==diam:
                count+=1
                if count>max_:
                    max_=count
                    max_node = node
    print("For {} the node with maximum connections having distance equal to that of the graph's diameter is node-{} with {} connections ".format(name,max_node,max_))
        

for graph,name in graph_dict.items():
    find_nodes(graph,name)
```

Στo REG δίκτυο η περιφέρεια και το κέντρο είναι τα ίδια κάθως όλοι οι κόμβοι έχουν την ίδια εκκεντρότητα.

Στα υπόλοιπα δίκτυα μας ενδιαφέρει όχι αναγκστικά να εξετάσουμε τους κόμμβους αυτούς καθ'αυτούς που ανήκουν στην περιφέρεια ή στο κέντρο,
αλλά το πλήθος των κόμβων. Στο Erdos-Renyi παρατηρούμε μια ισορροπία μεταξύ των κόμβων της περιφέρειας και του κέντρου μιας και σε αυτό το δίκτυο οι δύο ποσότητες είναι κοντά. 

Στην περίπτωση του RGG βλέπουμε πολύ λίγους κόμβους να ανήκουν στην περιφέρεια και στο κέντρο γεγονός που υποδεικνύει ότι
υπάρχουν πολλοί κόμβοι που πέφτουν στο ενδιάμεσο δηλαδή έχουν εκκεντρότητα radius < ε(v) < diameter.

Στα scale-free δίκτυα έχουμε τους κόμβους hubs στο κέντρο (που είναι λίγοι) και πάρα πολλους κόμβους στην περιφέρεια.

Ενώ τέλος στα SW παρατηρούμε μια άμβλυνση του φαινομένου των SF και απο την μεριά των κόμβων του κέντρου αλλά και της περιφέρειας.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    return [G.degree(n) for n in G.nodes()]

def plot_degree_distribution(G,name):
    degrees = find_degree(G)
    print("Mean degree for {} is {:.3f} with std {:.3f}".format(name,np.mean(degrees),np.std(degrees)))
    plt.hist(degrees)
    plt.xlabel("Degree")
    plt.title("Degree distribution for {}".format(name))
    plt.show()
    
for graph,name in graph_dict.items():
    plot_degree_distribution(graph,name)
```

#### Σχολιασμός

Για όλα τα παραπάνω δίκτυα δεν μπορούμε να είμαστε σίγουροι πόσο καλά προσεγγίζουν την θεωρητική τους κατανομή
μιας και βλέπουμε μόνο ένα τυχαίο Instance του δικτύου που μπορεί να αποτελεί κάποιου είδους εξαίρεσης.

Συγκεκριμένα, το Regular ακολουθεί επακριβώς την αναμενόμενη συμπεριφορά έχοντας μια κρουστική στο σημείο χ=d. Αναφορικά με τον μέσο βαθμό και την διακύμανση βλέπουμε πώς ο μέσος είναι ίσος με το d και η διακύμανση μηδενική.

Το erdos-renyi μοντέλο αναμένουμε να έχει δυωνιμική κατανομή επομένως θα μπορούσαμε να πούμε πως το σχήμα που βλέπουμε δεν είναι παράταιρο,
αλλά παρουσιάζει ένα ελαφρύ skewness. Η μέση τιμή είναι np (για το αντίστοιχο Gilbert Model) ενώ η διακύμανση npq αφού ακολουθεί διωνυμική κατανομή.

Στον Random Geometric Graph αναμένουμε να έχουμε μια κανονική κατανομή με πεπλατημένες ουρές πράγμα το οποίο δεν βλέπουμε εδώ, αυτό έχει
να κάνει τόσο με τις υπερπαραμέτρους του δικτύου καθώς και με την ίδια την απεικόνιση του ιστογράμματος (αριθμός bins). Εδώ ο μέσος του 
βαθμού είναι E(x) = (πρ^2/L^2) * n ~ 18.5. Όπως βλέπουμε είμαστε 1σ μακριά από τον αναμενόμενο μέσο.

Επακολούθως έχουμε το scale-free δίκτυο το οποίο ταιριάζει (εμπειρικά) πολύ καλά την αναμενόμενη (εκθετική) κατανομή. Γενικότερα ο
βαθμός αυτών των δικτύων ακολουθεί P(k) = κ^(-γ), γε[2.5,5.5] καθώς υπάρχει η αρχή του preferential attachment. Σε αυτή την περίπτωση μιας
και δεν γνωρίζουμε την τιμή του γ αλλά ξέρουμε οτι εν γένει ακολουθείται η power-law distribution που έχει μέση τιμή E(d)=(γ-1)/(γ-2) μπορούμε
να εκτιμήσουμε την τιμή του γ η οποία ανέρχεται στο 2.15.

Τέλος έχουμε το small-world δίκτυο στα οποία δίκτυα έχει μεγάλη σημασία το rewiring probability. Στην συγκεκριμένη περίπτωση 
αυτό ανέρχεται στο 0.25. Γενικά αναμένουμε μια καμπανοειδή καμπύλη δεξιά skewed. Στο συγκεκριμένο παράδειγμα παίρνουμε κάτι που προσεγγίζει

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    max_eigenval = nx.adjacency_spectrum(G).real.max()
    alpha = (1/max_eigenval) - 0.01
    return list(nx.katz_centrality(G,alpha).values())

def plot_katz_centrality_distribution(G,name):
    centralities = find_katz(G)
    print("Mean Katz-centrality for {} is {:.3f} with std {:.3f}".format(name, np.mean(centralities),np.std(centralities)))
    plt.hist(centralities)
    plt.xlabel("Centralities")
    plt.title("Katz  centrality distribution for {}".format(name))
    plt.show()
    
for graph,name in graph_dict.items():
    plot_katz_centrality_distribution(graph,name)
```

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G,name):
    max_degree_node = np.argmax(find_degree(G))
    max_closeness_node = np.argmax(find_closeness(G))
    max_betweenness_node = np.argmax(find_betweenness(G))
    max_katz_node = np.argmax(find_katz(G))
    print("For {} the node with the max: \nDegree: {}\nClossness: {}\nBetweenness: {}\nKatz: {}".format(name,max_degree_node,max_closeness_node,max_betweenness_node,max_katz_node))
    
    
for graph,name in graph_dict.items():
    find_central(graph,name)
```

```python
def find_clustering_coeff(G):
    return [nx.clustering(G, n) for n in G.nodes()]

def plot_clustering_coeff(G,name):
    clustering_coeffs = find_clustering_coeff(G)
    print("Mean degree for {} is {:.3f} with std {:.3f}".format(name,np.mean(clustering_coeffs),np.std(clustering_coeffs)))
    plt.hist(clustering_coeffs)
    plt.xlabel("Clustering coefficient")
    plt.title("Clustering coefficient distribution for {}".format(name))
    plt.show()

plot_clustering_coeff(G_got,"Game of thrones graph")
```

#### Ποιοι ήταν οι 5 κεντρικότεροι χαρακτήρες του πρώτου κύκλου επεισοδίων ως προς το βαθμό, την εγγύτητα και την ενδιαμεσικότητα ? 

```python
def find_top_central(G,name,k):
    top_degrees = np.argpartition(find_degree(G), -k)[-k:]
    top_closeness = np.argpartition(find_closeness(G), -k)[-k:]
    top_betweenness = np.argpartition(find_betweenness(G), -k)[-k:]
    
    node_labels = list(dict(G.nodes(data=True)).keys())
    
    print("Top five based on degree are {}".format([node_labels[i] for i in top_degrees]))
    print("Top five based on closeness are {}".format([node_labels[i] for i in top_closeness]))
    print("Top five based on betweenness are {}".format([node_labels[i] for i in top_betweenness]))

find_top_central(G_got,"aa",5)
```
